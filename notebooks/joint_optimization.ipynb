```python
import torch
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

from src.train_pendulum import train_joint
from src.util import set_seed, save_data

save_path = "/content/drive/MyDrive/Autonomous Control Systems/joint"

def pipe(seed=42, noise=0.0, save_path=""):

    set_seed(seed)

    agent, rewards, eval_rewards = train_joint(
        num_episodes=500,
        noise_std=noise,       # set to 0.1 for noisy variant
        device=device
    )

    if noise == 0.05:
        noise = 's'
    elif noise == 0.1:
        noise = 'm'
    elif noise == 0.3:
        noise = 'l'
    elif noise == 0:
        noise = 'na'

    save_data(
        {},
        seed=seed,
        group="joint",
        policy="td3",
        rewards=rewards,
        evals=eval_rewards,
        agent=agent,
        path=save_path,
        noise=noise

    )
```

```python
# seeds = [42, 123, 256, 789, 1024]
seeds = [3, 33, 333, 432, 3333]

env_noise = [0.05, 0.1, 0.3]

for seed in seeds:
    for noise in env_noise:
        pipe(seed, noise, save_path)
```
